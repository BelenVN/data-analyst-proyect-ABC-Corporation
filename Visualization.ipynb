```python
#Configuración. Primero me importo las librerias necesarias para el ejercicio

#Tratamiento de datos
# ---------------------------------------
import pandas as pd
import numpy as np

#Gestión de nulos 
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

#Visualización
#-----------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

#para configurar los dataframes y ver todas las columnas 
pd.set_option('display.max_columns', None)
```

```python
hr_data_clean = pd.read_csv('HR RAW DATA CLEAN.csv')
hr_data_clean
```

```python
hr_data_clean.columns
```

```python
#Let's compare the different levels of satisfactions
hr_data_satisfaction = hr_data_clean[["JobSatisfaction", "EnvironmentSatisfaction", "RelationshipSatisfaction"]]

hr_data_satisfaction
          
```

```python
Job_Satisfaction = hr_data_clean["JobSatisfaction"].value_counts(normalize=True)*100

Job_Satisfaction
```

```python
Environment_Satisfaction = hr_data_clean["EnvironmentSatisfaction"].value_counts(normalize=True)*100

Environment_Satisfaction
```

```python
Relationship_Satisfaction = hr_data_clean["RelationshipSatisfaction"].value_counts(normalize=True)*100

Relationship_Satisfaction
```

```python
# Crear una figura con una cuadrícula de 1 fila y 3 columnas
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

# Crear gráfico de pastel para Environment Satisfaction
axes[0].pie(Environment_Satisfaction,
            labels=["1", "2", "3", "4"], 
            autopct='%1.1f%%', 
            startangle=90)
axes[0].set_title("Environment Satisfaction")

# Crear gráfico de pastel para Relationship Satisfaction
axes[1].pie(Relationship_Satisfaction,
            labels=["1", "2", "3", "4"], 
            autopct='%1.1f%%', 
            startangle=90)
axes[1].set_title("Relationship Satisfaction")

# Crear gráfico de pastel para Job Satisfaction
axes[2].pie(Job_Satisfaction,
            labels=["1", "2", "3", "4"], 
            autopct='%1.1f%%', 
            startangle=90)
axes[2].set_title("Job Satisfaction")

# Ajustar el espacio entre subgráficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
# Mostrar valores únicos de cada columna
print('Valores únicos en EnvironmentSatisfaction:', hr_data_clean['EnvironmentSatisfaction'].unique())
print('Valores únicos en JobSatisfaction:', hr_data_clean['JobSatisfaction'].unique())
print('Valores únicos en RelationshipSatisfaction:', hr_data_clean['RelationshipSatisfaction'].unique())
```

```python
# Calcular la media de las tres columnas y crear una nueva columna 'SatisfactionOverall'
hr_data_clean['SatisfactionOverall'] = hr_data_clean[['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction']].mean(axis=1)

# Mostrar las primeras filas para verificar
print(hr_data_clean[['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction', 'SatisfactionOverall']].head())
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(12, 6))
sns.scatterplot(x='DistanceFromHome', y='SatisfactionOverall', hue='RemoteWork', data=hr_data_clean, palette='mako', alpha=0.7)

# Configurar los títulos y etiquetas
plt.title('Relación entre Satisfacción General y Distancia desde Casa')
plt.xlabel('Distancia desde Casa')
plt.ylabel('Satisfacción General')
plt.legend(title='Teletrabajo')
plt.grid(True)

plt.show()
```

```python
# Crear el gráfico de violín
plt.figure(figsize=(10, 6))
sns.violinplot(x='OverTime', y='WorkLifeBalance', data=hr_data_clean, palette='inferno')

# Configurar los títulos y etiquetas
plt.title('Distribución del Equilibrio entre Trabajo y Vida Personal según Horas Extras')
plt.xlabel('Horas Extras')
plt.ylabel('Equilibrio entre Trabajo y Vida Personal')
plt.grid(True)

plt.show()
```
