#### B) We note that we also have to homogenise the unique values of some columns

```python
# First: column MaritalStatus
# Custom function that evaluates the first character of the marital status
def categorize_status(status):
    # Check if the value is a string (str) before applying startswith
    if isinstance(status, str):
        if status.startswith(('m', 'M')):
            return 'Married'
        elif status.startswith(('d', 'D')):
            return 'Divorced'
        elif status.startswith(('s', 'S')):
            return 'Single'
    # If value is NaN or float, remains unchanged
    return status

# Apply the function to the column
df_clean['MaritalStatus'] = df_clean['MaritalStatus'].apply(categorize_status)

print(f'Value counts: {df_clean["MaritalStatus"].value_counts()}')
print(f'Value unique: {df_clean["MaritalStatus"].unique()}')
```

```python
# Secondly: column 'RemoteWork'
# Custom function that evaluates the first character of RemoteWork
def categorize_rwork(status):
    # Check if the value is a string (str) before applying startswith
    if isinstance(status, str):
        if status.startswith(('0')):
            return 'No'
        elif status.startswith(('1')):
            return 'Yes'
        elif status.startswith(('t', 'T')):
            return 'Yes'
        elif status.startswith(('f', 'F')):
            return 'No'
    # If value is NaN or float, remains unchanged
    return status

# Apply the function to the column
df_clean['RemoteWork'] = df_clean['RemoteWork'].apply(categorize_rwork)

df_clean.head(10)

```

```python
# Thirdly: column 'Gender'
# Custom function that evaluates the first character of 'Gender'
def categorize_gender(status):
    # Check if the value is a string (str) before applying '.startswith()'
    if isinstance(status, int):
        if status == 0:
            return 'Male'
        elif status == 1:
            return 'Female'
        
    # If value is NaN or float, remains unchanged
    return status

# Apply the function to the column
df_clean['Gender'] = df_clean['Gender'].apply(categorize_gender)

df_clean.head(10)
```

```python
# Filter rows where EmployeeNumber is not 'Unknown'
df_filtered = df_clean[df_clean['EmployeeNumber'] != 'Unknown']

# Filter rows with duplicate values in EmployeeNumber for further review
df_filtered_duplicates = df_filtered[df_filtered.duplicated(subset='EmployeeNumber', keep=False)]

# Show the first 59 rows of the sorted DataFrame
print(df_filtered_duplicates.to_string())
```

#### After reviewing our 208 duplicate rows (without 'Unknow' rows and after our Product Owner has discussed this situation with the customer, we need to do the following:
- For Remote Work, if there are discrepancies, take the value No.
- For Distance From Home, take the lower value or average the values if both are consistent.

```python
# Group by EmployeeNumber to apply rules
def resolve_duplicates(group):
    # For Remote Work: if there are discrepancies, take the value 'No'
    if group['RemoteWork'].nunique() > 1:
        remote_work_value = 'No'
    else:
        remote_work_value = group['RemoteWork'].iloc[0]
    
    # For Distance From Home: take the lowest value or the average if consistent
    if group['DistanceFromHome'].nunique() > 1:
        distance_value = group['DistanceFromHome'].min()  # Take the lowest value
    else:
        distance_value = group['DistanceFromHome'].mean()  # Or the average if they are consistent
    
    # Make sure DistanceFromHome is an integer
    distance_value = int(distance_value)
    
    # Return a merged row with set values
    return pd.Series({
        'EmployeeNumber': group['EmployeeNumber'].iloc[0],
        'RemoteWork': remote_work_value,
        'DistanceFromHome': distance_value,
        # Copy the rest of the columns
        **group.drop(columns=['RemoteWork', 'DistanceFromHome']).iloc[0]
    })

df_resolved = df_filtered_duplicates.groupby('EmployeeNumber').apply(resolve_duplicates).reset_index(drop=True)

df_resolved
```

##### Once we have managed the duplicate rows where the 'EmployeeNumber' is identified, let's concatenate it with the rest of the rows where the 'EmployeeNumber' is Unknown.

```python
# Apply the duplicate resolution function to each group
df_resolved = df_filtered_duplicates.groupby('EmployeeNumber').apply(resolve_duplicates).reset_index(drop=True)

# Filter rows where EmployeeNumber is 'Unknown'
df_unknown = df_clean[df_clean['EmployeeNumber'] == 'Unknown']

# Create DataFrame with original rows without duplicates and without 'Unknown'
df_original_no_duplicates = df_filtered.drop_duplicates(subset='EmployeeNumber', keep='first')

# Ensure that original unduplicated rows remain unchanged
df_original_no_duplicates = df_original_no_duplicates[~df_original_no_duplicates['EmployeeNumber'].isin(df_filtered_duplicates['EmployeeNumber'])]

# Merge original unduplicated rows, resolved duplicates, and rows with 'Unknown'
df_clean_final = pd.concat([df_original_no_duplicates, df_resolved, df_unknown], ignore_index=True)

# Show the entire merged DataFrame
print(f"Original unduplicated rows: {df_original_no_duplicates.shape[0]}")
print(f"Duplicate rows resolved: {df_resolved.shape[0]}")
print(f"Rows with 'Unknown': {df_unknown.shape[0]}")
print(f"Rows in df_clean_final: {df_clean_final.shape[0]}")

# Show a sample of the final DataFrame
print(df_clean_final.head())
print(df_clean_final.shape)



```

# PHASE 4 A/B TESTING
Determine if there is a relationship between the level of job satisfaction and employee turnover, and if so, determine the magnitude of that relationship.

```python
# We create a new column called "SatisfactionLevel" where the employees are divided between: Group A (Control): 
# Employees with a level of job satisfaction equal to or greater than 3 on a scale of 1 to 5 
# and Group B (Variant): Employees with a level of job satisfaction less than 3 on the same scale.

df_clean_final['SatisfactionLevel'] = df_clean_final['JobSatisfaction'].apply(lambda x: 'A' if x >= 3 else 'B')

# Show the DataFrame with the new column
df_clean_final
```

2. The turnover rate (percentage of employees who left the company) is calculated in each group.

```python
# Count the total number of employees in each group
total_employees = df_clean_final.groupby('SatisfactionLevel').size()

# Count the number of employees who left the company in each group
employees_who_left = df_clean_final[df_clean_final['Attrition'] == 'Yes'].groupby('SatisfactionLevel').size()

# Calculate the turnover rate (percentage of employees who left the company)
turnover_rate = (employees_who_left / total_employees) * 100

turnover_rate
```

3. A statistical analysis is performed to determine if there is a significant difference in the turnover rate between groups A and B.

```python
# Create a contingency table
contingency_table = pd.crosstab(df_clean_final['SatisfactionLevel'], df_clean_final['Attrition'])

# Perform the chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

print("\nChi-Square Test:")
print(f"Chi-cuadrado: {chi2}")
print(f"P-valor: {p_value}")

# Interpret the p-value
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in the turnover rate between the groups.")
else:
    print("There is no significant difference in turnover rate between the groups.")
```

5. Calculate the magnitude of this relationship using statistics such as the difference in means, for example.

```python
# Calculate the difference of means
median_difference = turnover_rate['B'] - turnover_rate['A']

print("Percentage of employees who have left the company in each group:")
print(turnover_rate)

# Show the difference of means
print(f"\nMean difference in turnover rate between groups:")
print(f"Difference: {median_difference:.2f}%")
```
